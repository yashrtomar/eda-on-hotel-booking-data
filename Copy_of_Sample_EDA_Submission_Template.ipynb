### Variables Description



1. Hotel: Resort Hotel or City Hotel

2. is_canceled: Value indicating if the booking was canceled (1) or not (0)

3. lead_time: Number of days between entering the date of the booking into the PMS and the arrival date

4. arrival_date_year: Year of arrival

5. arrival_date_month: Month of arrival

6. arrival_date_week_number: Week number of the year of arrival date

7. arrival_date_day_of_month: Day of arrival month

8. stays_in_weekend_nights: Number of weekend night stays or booked to stay

9. stays_in_week_nights: Number of week night stays or booked to stay

10. adults: Number of adults

11. children: Number of children

12. babies: Number of babies

13. meal: Meal type. Categories are presented in standard hospitality meal packages

14. country: Country of origin.

15. market_segment: Market segment designation. In categories, the term â€œTAâ€ means â€œTravel Agentsâ€ and â€œTOâ€ means â€œTour Operatorsâ€

16. distribution_channel: Booking distribution channel. The term â€œTAâ€ means â€œTravel Agentsâ€, â€œTOâ€ means â€œTour Operatorsâ€ and GDS means "Global Distribution System"

17. is_repeated_guest: Value indicating if the booking name was from a repeated guest (1) or not (0)

18. previous_cancellations: Number of previous bookings that were cancelled by the customer prior to the current booking

19. previous_bookings_not_canceled: Number of previous bookings not cancelled by the customer prior to the current booking

20. reserved_room_type: Code of room type reserved. Code is presented instead of designation for anonymity reasons.

21. assigned_room_type: Code for the type of room assigned to the booking.

22. booking_changes: Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation

23. deposit_type: Indication on if the customer made a deposit to guarantee the booking.

24. agent: ID of the travel agency that made the booking

25. company: ID of the company/entity that made the booking or responsible for paying the booking.

26. days_in_waiting_list: Number of days the booking was in the waiting list before it was confirmed to the customer

27. customer_type: Type of booking, assuming one of four categories

28. adr: Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights

29. required_car_parking_spaces: Number of car parking spaces required by the customer

30. total_of_special_requests: Number of special requests made by the customer (e.g. twin bed or high floor)

31. reservation_status: Reservation last status, assuming one of three categories

32. Canceled: booking was canceled by the customer

33. Check-Out: customer has checked in but already departed

34. No-Show: customer did not check-in and did inform the hotel of the reason why

35. reservation_status_date: Date at which the last status was set

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Drop the duplicates from the dataset
hotel_booking_df.drop_duplicates(inplace = True)

hotel_booking_df.shape

# Creating a duplicate of the original dataset before making any changes in it
hotel_booking_df_copy = hotel_booking_df.copy()
````

````python
hotel_booking_df_copy.columns
````

````python
# Replace null values in children column with 0 assuming that family had 0 children
hotel_booking_df_copy['children'].fillna(0 , inplace = True)

# Replace null values in company column with 0 assuming those rooms were booked without company
hotel_booking_df_copy['company'].fillna(0, inplace = True)

# Replace null values in agent columns with 0 assuming those rooms were booked without agent
hotel_booking_df_copy['agent'].fillna(0, inplace = True)

# Replace null values in country column as 'Others'
hotel_booking_df_copy['country'].fillna('others' , inplace = True)
````

````python
# Check for any null values after replacing them
hotel_booking_df_copy.isnull().sum()
````

````python
# Drop the column 'company' as it contains a lot of null values compared to other columns
hotel_booking_df_copy.drop(['company'], axis =1, inplace = True)
````

````python
# Drop rows with no adults

no_guest = hotel_booking_df_copy[hotel_booking_df_copy['adults'] + hotel_booking_df_copy['babies'] + hotel_booking_df_copy['children'] == 0]
hotel_booking_df_copy.drop(no_guest.index, inplace=True)
````

````python
# Add some new columns to make our data analysis ready

# Create total people column by adding all the members in that booking
hotel_booking_df_copy['total_members'] = hotel_booking_df_copy['adults'] + hotel_booking_df_copy['children'] + hotel_booking_df_copy['babies']

# Create a column to check the stay period by people in that booking
hotel_booking_df_copy['stay_period'] = hotel_booking_df_copy['stays_in_weekend_nights'] + hotel_booking_df_copy['stays_in_week_nights']
````
